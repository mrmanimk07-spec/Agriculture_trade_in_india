## Stage 2 – Data Cleaning and Pre-processing



**Changing datatypes**

First change the datatype of year from object to numeric because it contains null values afer converting the convert float to int datatype
Change the datatype of Export_Value_USD in object to numeric datatype

```python
df["Year"] = pd.to_numeric(df["Year"], errors='coerce')
df["Year"] = df["Year"].astype("Int64")
df["Export_Value_USD"] = pd.to_numeric(df["Export_Value_USD"], errors='coerce')
```

```python
df.dtypes
```

**Handling Missing Values for numerical columns**

```python
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
df.isnull().sum()
```

**Handling Misssing values for Categorial columns**

```python
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

**Treating Outliers Using Interquartile Range Method (IQR)**

```python
for col in ["Export_Quantity_Tonnes", "Import_Quantity_Tonnes", "Export_Value_USD", "Import_Value_USD","Tariff_Percent","Exchange_Rate"]:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] > upper, upper,
              np.where(df[col] < lower, lower, df[col]))
```

```python
df.shape
```

**Checking Skewness**

```python
df[num_cols].skew()
```

**Create four new Columns for Feature Transformation**

Here i am using pandas cut fuction for creating traffic category column using traffic percent column by breaking into bins(intervals)

```python
df["Trade_Balance"] = df["Export_Value_USD"] - df["Import_Value_USD"]
df["Export_Value_per_Tonne"] = df["Export_Value_USD"] / df["Export_Quantity_Tonnes"]
df["Import_Value_per_Tonne"] = df["Import_Value_USD"] / df["Import_Quantity_Tonnes"]
df["Tariff_Category"] = pd.cut(df["Tariff_Percent"],
                              bins=[0, 5, 15, 25],
                              labels=["Low", "Medium", "High"])
```
